```python
# Leverages numpy, SciKit-Fuzzy, and MatPlotLib
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe variables
x_closeness = np.arange(0, 101, 1)
x_speed     = np.arange(0, 101, 1)
x_change    = np.arange(-50, 51, 1) # Universe for the consequent

# Generate fuzzy membership functions for angle
closeness_vc = fuzz.trapmf(x_closeness, [0, 0, 25, 40])
closeness_c  = fuzz.trimf(x_closeness, [10, 30, 50])
closeness_cm = fuzz.trimf(x_closeness, [30, 50, 70])
closeness_f  = fuzz.trimf(x_closeness, [50, 70, 90])
closeness_vf = fuzz.trapmf(x_closeness, [60, 75, 100, 100])

# Visualize the universe and membership functions for angle
fig, ax0 = plt.subplots(figsize=(10, 5))

ax0.plot(x_closeness, closeness_vc, 'b', linewidth=1.5, label='Very Close')
ax0.plot(x_closeness, closeness_c, 'g', linewidth=1.5, label='Close')
ax0.plot(x_closeness, closeness_cm,  'r', linewidth=1.5, label='Comfortable')
ax0.plot(x_closeness, closeness_f, 'm', linewidth=1.5, label='Far')
ax0.plot(x_closeness, closeness_vf, 'c', linewidth=1.5, label='Very Far')
ax0.set_title('Closeness')
ax0.legend()

plt.plot()
```

```python
# Generate fuzzy membership functions
speed_slo = fuzz.trapmf(x_speed, [0, 0, 15, 50])
speed_med = fuzz.trimf(x_speed, [20, 50, 80])
speed_fas = fuzz.trapmf(x_speed, [50, 85, 100, 100])

# Visualize the universe and membership functions for speed
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(x_speed, speed_slo, 'r', linewidth=1.5, label='Very slow')
ax1.plot(x_speed, speed_med, 'g', linewidth=1.5, label='Slow')
ax1.plot(x_speed, speed_fas, 'b', linewidth=1.5, label='Fast')
ax1.set_title('Speed')
ax1.legend()

plt.plot()
```

```python
# Generate fuzzy membership functions
change_hin = fuzz.smf(x_change, 0, 50)
change_inc = fuzz.trimf(x_change, [0, 20, 40])
change_con = fuzz.trapmf(x_change, [-15, -5, 5, 15])
change_dec = fuzz.trimf(x_change, [-40, -20, 0])
change_hde = fuzz.zmf(x_change, -50, 0)

# Visualize the universe and membership functions for change
fig, ax2 = plt.subplots(figsize=(10, 5))

ax2.plot(x_change, change_hin, 'r', linewidth=1.5, label='Heavily Increase')
ax2.plot(x_change, change_inc, 'c', linewidth=1.5, label='Increase')
ax2.plot(x_change, change_con, 'g', linewidth=1.5, label='Constant')
ax2.plot(x_change, change_dec, 'm', linewidth=1.5, label='Decrease')
ax2.plot(x_change, change_hde, 'b', linewidth=1.5, label='Heavily Decrease')
ax2.set_title('Change (by how much to alter the set speed by)')
ax2.legend()

plt.plot()
```
