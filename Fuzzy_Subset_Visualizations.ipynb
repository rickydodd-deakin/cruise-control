```python
# Leverages numpy, SciKit-Fuzzy, and MatPlotLib
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe variables
x_angle = np.arange(-46, 47, 1)
# x_speed = np.arange(-100, 101, 1)
x_speed = np.arange(0, 101, 1)
x_change = np.arange(-100, 101, 1) # Universe for the consequent

# Generate fuzzy membership functions
angle_vd = fuzz.trimf(x_angle, [-46, -46, -23])
angle_sd = fuzz.trimf(x_angle, [-46, -23, 0])
angle_f  = fuzz.trimf(x_angle, [-23, 0, 23])
angle_su = fuzz.trimf(x_angle, [0, 23, 46])
angle_vu = fuzz.trimf(x_angle, [23, 46, 46])

# Visualize the universe and membership functions associated with said universe
fig, ax0 = plt.subplots(figsize=(10, 5))

ax0.plot(x_angle, angle_vd, 'b', linewidth=1.5, label='Very Downhill')
ax0.plot(x_angle, angle_sd, 'g', linewidth=1.5, label='Somewhat Downhill')
ax0.plot(x_angle, angle_f,  'r', linewidth=1.5, label='Flat')
ax0.plot(x_angle, angle_su, 'm', linewidth=1.5, label='Somewhat Uphill')
ax0.plot(x_angle, angle_vu, 'c', linewidth=1.5, label='Very Uphill')
ax0.set_title('Angle')
ax0.legend()

plt.plot()
```

```python
# Generate fuzzy membership functions
#speed_slo = fuzz.trimf(x_speed, [0, 0, 50])
speed_slo = fuzz.trapmf(x_speed, [0, 0, 15, 50])
speed_med = fuzz.trimf(x_speed, [20, 50, 80])
#speed_fas = fuzz.trimf(x_speed, [50, 100, 100])
speed_fas = fuzz.trapmf(x_speed, [50, 85, 100, 100])

# Visualize the universe and membership functions associated with said universe
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(x_speed, speed_slo, 'r', linewidth=1.5, label='Very slow')
ax1.plot(x_speed, speed_med, 'g', linewidth=1.5, label='Slow')
ax1.plot(x_speed, speed_fas, 'b', linewidth=1.5, label='Fast')
ax1.set_title('Speed')
ax1.legend()

plt.plot()
```

```python
# Generate fuzzy membership functions
change_inc = fuzz.smf(x_change, 0, 100)
change_con = fuzz.trapmf(x_change, [-50, -10, 10, 50])
change_dec = fuzz.zmf(x_change, -100, 0)

# Visualize the universe and membership functions associated with said universe
fig, ax2 = plt.subplots(figsize=(10, 5))

ax2.plot(x_change, speed_inc, 'r', linewidth=1.5, label='Increase')
ax2.plot(x_change, speed_con, 'g', linewidth=1.5, label='Constant')
ax2.plot(x_change, speed_dec, 'b', linewidth=1.5, label='Decrease')
ax2.set_title('Change (by how much to alter the set speed by)')
ax2.legend()

plt.plot()
```
