```python
# Leverages numpy, SciKit-Fuzzy, and MatPlotLib
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe variables
x_angle = np.arange(-46, 47, 1)
x_speed = np.arange(40, 101, 1)

# Generate fuzzy membership functions
angle_vd = fuzz.trimf(x_angle, [-46, -46, -23])
angle_sd = fuzz.trimf(x_angle, [-46, -23, 0])
angle_f  = fuzz.trimf(x_angle, [-23, 0, 23])
angle_su = fuzz.trimf(x_angle, [0, 23, 46])
angle_vu = fuzz.trimf(x_angle, [23, 46, 46])

# Visualize the universe and membership functions associated with said universe
fig, ax0 = plt.subplots(figsize=(10, 5))

ax0.plot(x_angle, angle_vd, 'b', linewidth=1.5, label='Very Downhill')
ax0.plot(x_angle, angle_sd, 'g', linewidth=1.5, label='Somewhat Downhill')
ax0.plot(x_angle, angle_f,  'r', linewidth=1.5, label='Flat')
ax0.plot(x_angle, angle_su, 'm', linewidth=1.5, label='Somewhat Uphill')
ax0.plot(x_angle, angle_vu, 'c', linewidth=1.5, label='Very Uphill')
ax0.set_title('Angle')
ax0.legend()

ax0.spines['top'].set_visible(False)
ax0.spines['right'].set_visible(False)
ax0.get_xaxis().tick_bottom()
ax0.get_yaxis().tick_left()

plt.plot()
```

```python
# Generate universe variables
x_angle = np.arange(-46, 47, 1)
x_speed = np.arange(-100, 101, 1)
x_actualAcceleration  = np.arange(0, 151, 1)

# Generate fuzzy membership functions
speed_inc = fuzz.smf(x_speed, 0, 100)
speed_con = fuzz.trapmf(x_speed, [-25, -5, 5, 25])
speed_dec = fuzz.zmf(x_speed, -100, 0)

# Visualize the universe and membership functions associated with said universe
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(x_speed, speed_inc, 'r', linewidth=1.5, label='Increase')
ax1.plot(x_speed, speed_con, 'g', linewidth=1.5, label='Constant')
ax1.plot(x_speed, speed_dec, 'b', linewidth=1.5, label='Decrease')
ax1.set_title('Speed')
ax1.legend()

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.get_xaxis().tick_bottom()
ax1.get_yaxis().tick_left()

plt.plot()
```
